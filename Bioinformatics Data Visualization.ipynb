```python
from __future__ import print_function
import warnings
import pytraj as pt
import nglview as nv
from IPython.display import IFrame

warnings.filterwarnings('ignore') # make the notebook nicer

print("nglview version = {}".format(nv.__version__))
print("pytraj version = {}".format(pt.__version__))

traj = pt.load('6w4b.pdb')
view = nv.show_pytraj(traj)
view
```

```python
view.clear_representations()
#add ribbons
view.add_cartoon('not protein')
#add ball and stick for non-rotien
view.add_ball_and_stick('protein')
view
```

```python
from Bio.PDB import *
parser = MMCIFParser()
structure = parser.get_structure('6w4b', 'w4/6w4b.cif')
for model in structure:
    print(f"model {model}")
```

```python
#Ok we have one model.
#But we should have several chains as we saw above (2 symetrical chaine (avec un axe de symetrie)).
model = structure[0] #since we only have one model
for chain in model:
    print(f"chain {chain}, Chain ID: {chain.id}")
```

```python
#We have 2 chains: A, and B.
#We can go one step further and get all the residues in a chain.
#We can access individual chains like keys in a dictionary from a model.
chain_A = model['A']
for res in chain_A:
    print(f"Residue name: {res.resname}, number: {res.id[1]}")
```

```python
# Get number of atome per chain
p_6w4b = structure 
def describe_model(name, pdb):
    print()
    for model in p_6w4b:
        for chain in model:
            print('%s - Chain: %s. Number of residues: %d. Number of atoms: %d.' %
                  (name, chain.id, len(chain), len(list(chain.get_atoms()))))
describe_model('6w4b', p_6w4b)
```

```python
#Now we see that each chain is made up of Residue objects.
#Finally, each residue should be made up of atoms.
#Let's get residue 10 from Chain A.
res = chain_A[10]
print("Residue name: LEU, number: 10",res)
```

```python
for atom in res:
    print(f"{atom.name}")
```

```python
#Putting it all together, we can print every atom in every chain.
for model in structure:
    for chain in model:
        for residue in chain:
            for atom in residue:
                print(atom)
```

> find all other residues within a certain distance cutoff

> distance cutoff of 10 Angstroms.

```python
residue_1_CA = structure[0]['A'][10]['CA']
# le resultat est un vecteur
print("1er residu coordonnee du vecteur", residue_1_CA.coord)
#Now we can use this to compute distances!
#Let's take another atom.
residue_2_CA = structure[0]['A'][30]['CA']
# le resultat est un vecteur
print("2eme residu coordonnee du vecteur", residue_2_CA.coord)
Diff = residue_1_CA - residue_2_CA
print("Distance", Diff)
```
